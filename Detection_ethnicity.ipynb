```python
import matplotlib.pyplot as plt
from detecto.utils import read_image
```

```python
data_path="/content/customimages/MyDrive/customimages/"
sample_image = "people1.jpg"
image = read_image(data_path+sample_image)
plt.imshow(image)
plt.show()
```

```python
from detecto.core import Dataset
dataset = Dataset(data_path)
```

```python
from detecto.visualize import show_labeled_image

image, targets = dataset[0]

show_labeled_image(image, targets['boxes'], targets['labels'])
```

```python
from detecto.core import DataLoader, Model

labels = ['American','African','Indian','Asian']

model = Model(labels)
```

```python
model.fit(dataset)
```

```python
image = read_image(data_path+ 'people1.jpg')
labels,boxes,scores = model.predict(image)
```

```python
print("labels",labels)
print("boxes",boxes)
print("scores",scores)
```

```python
from detecto.visualize import show_labeled_image

for i in range(10):
      show_labeled_image(image, boxes[i], labels[i])
```
